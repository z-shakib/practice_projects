```python
# Import dependencis 

import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler 
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('diabetes.csv') 
dataset.head() 
```

```python
dataset.shape
```

```python
# Statestical Summary 

dataset.describe() 
```

```python
dataset.info() 
```

```python
# check the balance of outcoume

dataset['Outcome'].value_counts() 
```

```python
# Separate X and Y 

X = dataset.drop(['Outcome'], axis = 1) 
Y = dataset['Outcome'] 
```

```python
X
```

```python
# Split Train and Test

X_train, x_test, Y_train, y_test = train_test_split(
    X, Y, test_size = 0.2, stratify=Y, random_state = 6
)
```

```python
print(X.shape, X_train.shape, x_test.shape)
```

```python
#! MAke an oobject of standardscaler 

scaler = StandardScaler() 
```

```python
# standardize train and test data

X_train = scaler.fit_transform(X_train) 
x_test = scaler.transform(x_test)
```

```python
X_train 
```

```python
x_test
```

```python
# create an object of svm algorithm 

classifier = svm.SVC(kernel = 'linear') 
```

```python
# now train the model 

classifier.fit(X_train, Y_train) 
```

```python
# check the training accuracy 

X_train_prediction = classifier.predict(X_train) 
train_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
```

```python
print('Accuracy On Traing Data: ', train_data_accuracy)
```

```python
# check the test accuracy

x_test_prediction = classifier.predict(x_test) 
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy On test data: ', test_data_accuracy) 
```

```python
# defining an function to predict diabetic 
columns = [
    'Pregnancies',
    'Glucose',
    'BloodPressure',
    'SkinThickness',
    'Insulin',
    'BMI',
    'DiabetesPedigreeFunction',
    'Age'
]


def predict_diabetes(new_patient):
    # convert to DataFrame 
    df = pd.DataFrame([new_patient], columns=columns)
    # standardize input data
    std_data = scaler.transform(df)
    # prediction
    prediction = classifier.predict(std_data)

    if prediction[0] == 0:
        return "Patient is Not Diabetic"
    else:
        return "Patient is Diabetic"
```
